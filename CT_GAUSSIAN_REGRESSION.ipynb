```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel
```

```python
kernel = DotProduct() + WhiteKernel()
gpr = GaussianProcessRegressor(kernel=kernel,random_state=0).fit(X_train_set, Y_train_set)
gpr.score(X_train_set, Y_train_set)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
Predict = gpr.predict(X_train_set)
lin_mse = mean_squared_error(Y_train_set, Predict) 
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import precision_score, recall_score
precision_score(Y_test_set, Predict) # == 4096 / (4096 + 1522) 0.7290850836596654
recall_score(Y_test_set, Predict) 
```
